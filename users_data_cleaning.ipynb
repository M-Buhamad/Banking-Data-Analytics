```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("users_data.csv",dtype={'gender':'category'})
```

```python
df.info()
```

```python
df.set_index('id',inplace=True)
```

```python
df['per_capita_income'] = df['per_capita_income'].str.replace('$', '', regex=False).astype(int)
```

```python
df['yearly_income'] = df['yearly_income'].str.replace('$', '', regex=False).astype(int)
```

```python
df['total_debt'] = df['total_debt'].str.replace('$', '', regex=False).astype(int)
```

```python
df.info()
```

```python
df.loc[df.isna().any(axis=1)] # NO row has a missing value NaN
```

```python
df[df['birth_month']> 12] # No wrong entries
```

```python
df[df['retirement_age']> 80] # No wrong entries
```

```python
df.loc[(2019 - df['birth_year'] > df['current_age']+1)] # no wrong entry of age and birth year
```
